```python
import pandas as pd
data = pd.read_csv('CC GENERAL.csv', sep = ',')
data
```

```python
desc = data["PURCHASES"].describe()
desc
```

```python
data["PURCHASES"] = data["PURCHASES"].astype(int)
```

```python

#data.loc[(data["PURCHASES"] >=1003) & (data["PURCHASES"]) < 10000,"label" ]=2
#data.loc[(data["PURCHASES"] >= 10000) & (data["PURCHASES"]) <24000,"label"]  = 3
#data.loc[(data["PURCHASES"] >=24000) & (data["PURCHASES"]) < 49040,"label"]=4
for i ,row in data.iterrows():
    if row["PURCHASES"] >=0 and row["PURCHASES"] <=10000:
        data.loc[i,"label"] = 0
    elif row["PURCHASES"] >10000 and row["PURCHASES"] <=20000:
        data.loc[i,"label"] = 1
    elif row["PURCHASES"] >20000 and row["PURCHASES"] <=30000:
        data.loc[i,"label"] = 2
    elif row["PURCHASES"] >30000 and row["PURCHASES"] <=40000:
        data.loc[i,'label'] = 3
    elif row["PURCHASES"] >40000 and row["PURCHASES"] <49040:
        data.loc[i,'label'] = 4
    else:
        data.loc[i,'label'] = 5
    
data.head()
```

```python
data["label"].unique()
```

```python
data.isnull().sum()
data.drop(columns = "CUST_ID", inplace = True)
```

```python
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean(), inplace = True)

data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean(), inplace = True)
```

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(data)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data['BALANCE'] , data["PURCHASES"] , c= agglomerative[0], s=30)
ax.set_title("BALANCE AND PURCHASES")
ax.set_xlabel("BALANCE")
ax.set_ylabel("PURCHASES")
plt.colorbar(scatter)
```
